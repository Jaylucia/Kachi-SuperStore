**EXECUTIVE STORY & RECOMMENDATIONS**

After thorough analysis, it is eveident that the company has a very good sales performance. Despite strong sales performance, the company is struggling with profitability. Overall profit margin is approximately 12%, which is weak for a retail business and indicates inefficiencies in *pricing, discount strategy*, and *regional performance*.



Key findings from the analysis show:

*   Higher discounts lead to lower profits and even negative profit margins, meaning discounts are eroding revenue instead of driving healthy growth.
*   The Central region records strong sales but suffers from the worst profit performance, including negative profit margins, especially in Furniture and Office Supplies.
*   Some products generate consistent losses, while others only sell when heavily discounted, suggesting poor product pricing and portfolio management.
*   Shipping mode also impacts profitability, with Second Class shipping being the most profitable and Standard Class the least profitable.
*   Customer repeat rate is very high (~98%), meaning demand is strong, but the company is failing to convert this demand into healthy profits.


Overall, the business problem is not lack of sales, it is inefficient revenue quality caused by:

* Excessive discounting;

* Loss-making products;

* Poor regional profitability;

* Inefficient shipping strategies.


**RECOMMENDATIONS (Actionable Business Steps)**

**1. Fix Discount Strategy**

The problem of high discounts causing negative profit margins was detected, evidenced in higher discount levels leading to lower profit and negative margins. I recommend that the Management cap discounts at 20% except for controlled promotions. They should also stop blanket discounting and apply discounts only to overstocked products and low-demand SKUs. They can also introduce product-level discount rules instead of store-wide discounts.

**2. Address Central Region Losses**

The Central region has good sales but has the worst profitability as I detected negative profit margin, especially in Furniture and Office Supplies. I recommend that they audit Central region operations, especially the shipping costs, product mix, and discount levels.The Management should also consider reducing focus on Furniture and Office Supplies in Central region ans shift sales focus toward Technology category, which performs better.


**3. Remove or Reprice Loss-Making Products**

Some products like Cubify CubeX printers, Zebra GK420t, and  certain Lexmark printers consistently generate losses. These products should be discontinued as well as other consistently loss-making products. They can also consider increasing prices for those products while reviewing supplier costs for these SKUs.

**4. Improve Product Portfolio Strategy**

Based on the analysis, it is evident that some products like Ibico EPK-21 Electric Binding System only sell when discounted. The Management team should treat these as clearance products or reposition as premium items with better margins. They do not need to rely on discounts to drive volume.

**5. Optimize Shipping Strategy**

Acording to the investigation, it is evident that shipping mode affects profit. It was also detected that Second Class is the most profitable while Standard Class is the least profitable. The Management should encourage customers toward Second Class shipping while reviewing logistics contracts for Standard Class delivery. They should also reduce expensive shipping options for low-margin orders.




**FINAL BUSINESS CONCLUSION**

The company is experiencing strong demand but weak profitability due to **excessive discounting, loss-making products**, and **poor regional performance**. Strategic reduction of discounts, removal of unprofitable products, and targeted regional improvements, especially in the Central region, can significantly improve overall profit margins without reducing sales volume.
